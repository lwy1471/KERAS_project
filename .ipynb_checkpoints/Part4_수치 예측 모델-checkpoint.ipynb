```python
#선형회귀 모델
import numpy as np
from sklearn.metrics import mean_squared_error
import random

x_train = np.random.random((1000, 1))
y_train = x_train * 2 + np.random.random((1000,1)) / 3.0
x_test = np.random.random((100,1))
y_test = x_test * 2 + np.random.random((100,1)) / 3.0


x_train = x_train.reshape(1000,)
y_train = y_train.reshape(1000,)
x_test = x_test.reshape(100,)
y_test = y_test.reshape(100,)

%matplotlib inline
import matplotlib.pyplot as plt

plt.plot(x_train, y_train, 'ro')
plt.plot(x_test, y_test, 'bo')
plt.legend(['train','test'], loc='upper left')
plt.show()


w = np.cov(x_train, y_train, bias=1)[0,1] / np.var(x_train)
b = np.average(y_train) - w*np.average(x_train)

print("W:",w,", B:", b)

y_predict = w*x_test + b
print("y_predict")
print(y_predict)
mse = mean_squared_error(y_test, y_predict)
print("mse : " + str(mse))
```

```python
#퍼셉트론 신경망
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
import random

x_train = np.random.random((1000, 1))
y_train = x_train * 2 + np.random.random((1000,1)) / 3.0
x_test = np.random.random((100,1))
y_test = x_test * 2 + np.random.random((100,1)) / 3.0


%matplotlib inline
import matplotlib.pyplot as plt

plt.plot(x_train, y_train, 'ro')
plt.plot(x_test, y_test, 'bo')
plt.legend(['train','test'], loc='upper left')
plt.show()


model = Sequential()
model.add(Dense(1, input_dim=1))

model.compile(optimizer='rmsprop', loss='mse')

hist = model.fit(x_train, y_train, epochs=50, batch_size=64)
w, b = model.get_weights()


print ("W: ",w,", B: ",b)

plt.plot(hist.history['loss'])
plt.ylim(0.0, 3.5)
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train'], loc='upper left')
plt.show()

loss = model.evaluate(x_test, y_test, batch_size=32)
print('loss : ' + str(loss))
```

```python
#다층퍼셉트론 신경망
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
import random

x_train = np.random.random((1000, 1))
y_train = x_train * 2 + np.random.random((1000,1)) / 3.0
x_test = np.random.random((100,1))
y_test = x_test * 2 + np.random.random((100,1)) / 3.0


%matplotlib inline
import matplotlib.pyplot as plt

plt.plot(x_train, y_train, 'ro')
plt.plot(x_test, y_test, 'bo')
plt.legend(['train','test'], loc='upper left')
plt.show()


model = Sequential()
model.add(Dense(64, input_dim=1, activation='relu'))
model.add(Dense(1))

model.compile(optimizer='rmsprop', loss='mse')

hist = model.fit(x_train, y_train, epochs=50, batch_size=64)

plt.plot(hist.history['loss'])
plt.ylim(0.0, 1.5)
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train'], loc='upper left')
plt.show()

loss = model.evaluate(x_test, y_test, batch_size=32)
print('loss : ' + str(loss))
```

```python
#깊은 다층퍼셉트론 신경망 모델
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
import random

x_train = np.random.random((1000, 1))
y_train = x_train * 2 + np.random.random((1000,1)) / 3.0
x_test = np.random.random((100,1))
y_test = x_test * 2 + np.random.random((100,1)) / 3.0


%matplotlib inline
import matplotlib.pyplot as plt

plt.plot(x_train, y_train, 'ro')
plt.plot(x_test, y_test, 'bo')
plt.legend(['train','test'], loc='upper left')
plt.show()


model = Sequential()
model.add(Dense(64, input_dim=1, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(1))

model.compile(optimizer='rmsprop', loss='mse')

hist = model.fit(x_train, y_train, epochs=50, batch_size=64)


plt.plot(hist.history['loss'])
plt.ylim(0.0, 1.5)
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train'], loc='upper left')
plt.show()

loss = model.evaluate(x_test, y_test, batch_size=32)
print('loss : ' + str(loss))
```
